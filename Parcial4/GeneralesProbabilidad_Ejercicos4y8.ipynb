```python
#Graficar P(n<=80) en función de n
dias = 365
personas = np.arange(1,81)
grafica = np.zeros(80)
probabilidad = np.zeros(365)

datos = []
for i in range(len(probabilidad)):
    x = (dias - i)/dias
    datos.append(x)
    if i <= 80:
        grafica[i - 1] = np.prod(datos)
        probabilidad[i - 1] = np.prod(datos)
    else: 
        probabilidad[i - 1] = np.prod(datos)


plt.scatter(personas,grafica)
plt.xlabel("Número de personas")
plt.ylabel("Probabilidad")
plt.title("Probabilidad de que las personas tengan fechas de cumpleaños diferentes")
```

Ejercicio 8

```python
#Hallar la probabilidad de obtener dos caras y dos sellos
def GetSample(N=int(1e5), ncoins=4):
    Sample = np.random.choice([1, -1], size=(N, ncoins), p=[0.5, 0.5])
    return Sample

N = int(1e5)
Sample = GetSample(N, ncoins=4)
contador = np.sum(np.sum(Sample == 1, axis=1) == 2)
probabilidad = contador / N

print("La probabilidad de obtener dos caras y dos sellos es:", probabilidad)
```
